### **`Observation`**

### **`Possible Reasons`**
- **UPI convenience, instant bank transfers, and widespread smartphone adoption may have accelerated P2P payment dominance.**
- **Increasing merchant onboarding, QR code adoption, and post-pandemic digital payment behavior likely boosted merchant transactions.**

```python
plot_atv_by_transaction_type(agg_txn)
```

### **`Observation`**
- **Peer-to-peer (P2P) payments show the highest average transaction value (ATV) among all transaction categories.**
- **Merchant payments have the lowest ATV, indicating frequent smaller-value transactions in daily commercial activities.**

### **`Interpretation`**
- **Users primarily rely on P2P payments for larger-value money transfers such as personal transfers, rent, and financial sharing activities.**
- **Lower ATV in merchant payments suggests PhonePe is widely used for routine small-value purchases and QR-based retail payments.**

### **`Business Meaning`**
- **High P2P ATV highlights strong user trust in PhonePe for direct financial transactions and high-value UPI transfers.**
- **Low-value high-frequency merchant payments indicate deep digital payment penetration in everyday consumer behavior.**

### **`Possible Reasons`**
- **Instant bank transfers and UPI convenience may have encouraged larger-value peer-to-peer transactions.**
- **Widespread QR adoption in grocery stores, cafes, and local merchants likely increased small-value merchant payment activity.**

### **`Business Meaning`**

### **`Possible Reasons`**
