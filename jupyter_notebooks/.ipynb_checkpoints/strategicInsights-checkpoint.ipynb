```python
plot_potential_user_segments(map_user)
```

### **`Observation`**
- **Several metropolitan districts such as Bengaluru Urban, Pune, and Thane show very high registered user counts but relatively lower engagement levels.**
- **Major urban regions dominate the list of high-user but low-engagement districts.**

### **`Interpretation`**
- **Many users in these districts have adopted PhonePe, but app activity per user remains comparatively lower.**
- **These regions may contain partially active or less-retained users despite having a strong user base.**

### **`Business Meaning`**
- **These districts represent high-potential reactivation markets where improving engagement could significantly increase platform activity.**
- **Targeted campaigns, merchant offers, and personalized financial services in these regions may improve long-term user retention and engagement.**

### **`Possible Reasons`**
- **Users in metro regions may use multiple competing payment platforms, reducing overall PhonePe engagement frequency.**
- **Large registered user bases naturally reduce engagement ratios when app usage growth does not match user acquisition growth.**

```python
plot_potential_growth_regions(map_user)
```

### **`Observation`**
- **Districts such as Kamle, Senapati, and North Garo Hills show exceptionally high PhonePe user growth percentages over time.**
- **Most potential growth regions belong to northeastern and digitally emerging districts with previously low adoption levels.**

### **`Interpretation`**
- **These districts are rapidly transitioning from low digital penetration to active PhonePe adoption and fintech participation.**
- **The strong growth trend indicates increasing awareness and acceptance of digital payment systems in emerging regions.**

### **`Business Meaning`**
- **These districts represent future high-potential markets for expanding PhonePe services and financial products.**
- **Early growth in underserved regions creates opportunities for long-term customer acquisition and deeper digital ecosystem expansion.**

### **`Possible Reasons`**
- **Improved internet connectivity, smartphone accessibility, and UPI awareness may have accelerated adoption in these districts.**
- **Lower initial user bases naturally resulted in very high percentage growth rates over time.**

```python
plot_unusual_transaction_patterns(map_map)
```

### **`Observation`**
- **States such as Karnataka, Maharashtra, Telangana, and Andhra Pradesh show very high transaction variability and wider transaction amount distributions.**
- **Several states contain extreme transaction ranges and noticeable outliers, indicating inconsistent transaction behavior over time.**

### **`Interpretation`**
- **These states experience highly dynamic digital payment activity with both very large and moderate transaction volumes.**
- **Wide transaction distributions may indicate seasonal spikes, high business activity, or sudden surges in digital payment usage.**

### **`Business Meaning`**
- **High transaction variability states represent economically active digital payment markets with strong fintech dependency.**
- **Identifying unusual transaction patterns can help monitor changing payment behavior, operational trends, and potential risk regions.**

### **`Possible Reasons`**
- **Large urban populations, business ecosystems, and high merchant activity may contribute to fluctuating transaction amounts.**
- **Festive seasons, business payments, and rapid increases in UPI adoption may create temporary transaction spikes and outliers.**
